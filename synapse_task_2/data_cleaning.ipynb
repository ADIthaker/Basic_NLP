```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
```

```python
with open('./data.json') as f:
  data = json.load(f)
data['data']
```

```python
df = pd.io.json.json_normalize(data['data'])
df['client_time'] = pd.to_datetime(df['client_time'],unit='s')
df
```

```python
category_sports = (df['category']=='Sports') | (df['category']=='Environment')  
df_category = df[category_sports]
df_category
```

```python
df1 = df_category.copy(deep=True)
df1['amount'] = df1['amount'].fillna(0)
df1
```

```python
amt = (df1['amount']==0) | (df1['amount']>=20)
df1 = df1[amt]
df1
```

# Inferences about ages

```python
df1_groupby_age = df1.groupby(by=df1['age'])['amount'].mean()
#print(df1_groupby_age)
ages_count = df1['age'].value_counts().count
age_groups = df1['age'].value_counts().index.tolist()
#print(ages)
plt.plot(age_groups,df1_groupby_age,label="Age")
plt.legend()
plt.title('Mean amount funded by age groups')
ymin, ymax = plt.ylim()

plt.ylim(0, ymax*1.1)
plt.show()

```

# Inferences on marital status

```python
df1_groupby_age = df1.groupby(by=df1['marital_status'])['amount'].mean()
#print(df1_groupby_age)
ages_count = df1['marital_status'].value_counts().count
age_groups = df1['marital_status'].value_counts().index.tolist()
#print(ages)
plt.bar(age_groups,df1_groupby_age,label="Marital Status")
plt.legend()
plt.title('Mean amount funded by marital status')
ymin, ymax = plt.ylim()

plt.ylim(0, ymax*1.1)
plt.show()
```

# Inferences on location state

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(15, 5), dpi=80, facecolor='w', edgecolor='k')
df1_groupby_state = df1.groupby(by=df1['location.state'])['amount'].mean()
#print(df1_groupby_age)
states_count = df1['location.state'].value_counts().count
states_groups = df1['location.state'].value_counts().index.tolist()

fig, ax = plt.subplots(figsize=(15, 5))

plt.bar(states_groups,df1_groupby_state)
plt.xticks(rotation='vertical')
for i, v in enumerate(df1_groupby_state):
    ax.text(i-.35, v+.5, str(round(v)), color='orange', fontweight='bold')
plt.xlabel('States')
plt.ylabel('Average amount funded in USD')
plt.title('Mean amount funded by states')
ymin, ymax = plt.ylim()

plt.ylim(0, ymax*1.1)


plt.show()
```
